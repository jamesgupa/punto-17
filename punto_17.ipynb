```python
#Ejercicio 17
import math

class Calculos:
  @staticmethod
  def longitud_circunferencia(radio):
    return 2 * math.pi * radio

  @staticmethod
  def area_circulo(radio):
    return math.pi * math.pow(radio,2)

radio = 9

longitud_circunferencia = Calculos.longitud_circunferencia(radio)
area_circulo = Calculos.area_circulo(radio)

print("Longitud circunferencia: ", longitud_circunferencia)
print("Área circulo: ", area_circulo)
```
